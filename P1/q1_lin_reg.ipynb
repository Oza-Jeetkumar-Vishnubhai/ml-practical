```python
import sklearn.metrics as sm
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
```

```python
# load csv and colums
df = pd.read_csv("iris_dataset.csv")

input_cols = ['Sepal.Width', 'Petal.Length', 'Petal.Width']
output_cols = ['Sepal.Length']
x = df[input_cols]
y = df[output_cols]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.41, shuffle=False)

lin_reg = linear_model.LinearRegression()
model = lin_reg.fit(x_train, y_train)

print(model.intercept_)
print(model.coef_)
```

```python
f, axs = plt.subplots(2, 2, figsize=(10, 7))

#Graph-1
G1 = sns.regplot(data=df, x="Sepal.Width", y="Sepal.Length", ax = axs[0][0], color = 'blue')
G1.set(xlabel = 'Sepal Width', ylabel = 'Sepal Length')
G1.set_title('Sepal Width v/s Sepal Length',color = 'blue')

#Graph-2
G2 = sns.regplot(data=df, x="Petal.Length", y="Sepal.Length", ax = axs[0][1], color ='red')
G2.set(xlabel = 'Petal Length', ylabel = 'Sepal Length')
G2.set_title('Petal Length v/s Sepal Length',color = 'red')


#Graph-3
G3 = sns.regplot(data=df, x="Petal.Width", y="Sepal.Length", ax = axs[1][0], color ='black')
G3.set(xlabel = 'Petal Width', ylabel = 'Sepal Length')
G3.set_title('Petal Width v/s Sepal Length',color = 'black')

#Graph-4
G3 = sns.regplot(data=df, x="Sepal.Length", y="Petal.Width", ax = axs[1][1], color ='pink')
G3.set(xlabel = 'Petal Width', ylabel = 'Sepal Length')
G3.set_title('Petal Width v/s Sepal Length',color = 'yellow')



f.tight_layout()
```
