```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("abalone.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Sex']=df['Sex'].astype('category').cat.codes
df.head()
```

```python
X = df.iloc[0:,1:]
y = df['Sex']
X.head()
```

```python
X.describe()
```

```python
set(y)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(X)
X_transform=ss.transform(X)
```

```python
pd.DataFrame(X_transform,columns=X.columns)
pd.DataFrame(X_transform,columns=X.columns).describe()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_transform,y,random_state=42,test_size=0.2)
```

```python
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred=log.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
y_proba=log.predict_proba(X_test)
y_proba
```
